## <hr class="solid"> 1. Housekeeping <a id='sec1'></a>

```python
# Importing necessary libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import constants
```

```python
# Reading input file

data = pd.read_csv('Weather Data.csv')
data
```

## <hr class="solid">2. Analyzing the Continuous Value - Temperature <a id='sec2'></a>

### 2.1. General Statistical Analysis

```python
# using pandas
min_value = data['Temp_C'].min()
max_value = data['Temp_C'].max()
quantiles = data['Temp_C'].quantile([0.25, 0.5, 0.75])

# using numpy
mean = np.mean(data['Temp_C'])
median = np.median(data['Temp_C'])
std_dev = np.std(data['Temp_C'])

print(f"Minimum Temperature: {min_value}")
print(f"Maximum Temperature: {max_value}")
print(f"Mean Temperature: {mean}")
print(f"Median Temperature: {median}")
print(f"Standard Deviation in Temperature: {std_dev}")
print("Quantiles:")
print(quantiles)
```

### 2.2. Analyzing the Temperature

#### 2.2.1. Understanding the trend of the Temperature recored

```python
data['Dt'] = pd.to_datetime(data['Date/Time'])
plt.figure(figsize=(15, 6))
plt.plot(data.index, data['Temp_C'], color='green')
plt.xlabel('Date/Time')
plt.ylabel('Temperature')
plt.title('Temperature Time Series Plot')
plt.show()
```

#### 2.2.2. Understanding the frequency distribution of the Temperature recored

```python
plt.figure(figsize=(15, 6))
plt.hist(data['Temp_C'], bins=20, color='green', alpha=0.7)
plt.xlabel('Tempareture')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.tight_layout()
plt.show()
```

### 2.3. Finding Outliers Values

<H4> Using Z-Score

```python
data['Z_Score'] = (data['Temp_C'] - data['Temp_C'].mean()) / data['Temp_C'].std()
threshold = 5
outliers = data[abs(data['Z_Score']) > threshold]
print("Outliers identified using z-score method:")
outliers
```

<H4> Using BoxPlot

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data['Temp_C'], patch_artist=True, boxprops=dict(facecolor='green'))
plt.xlabel('Temperature')
plt.ylabel('Value')
plt.title('Box Plot of Temperature Data')
plt.show()
```

## <hr class="solid"> 4. Relationship between continuous values  <a id='sec4'></a>

### 4.1. Continuous Values to be Analyzed

<h4> 1. Temperature </br>
2. Humidity </h4>

### 4.2. Science behind the relationship of these continuous values

<H4> The relationship between temperature and humidity is governed by thermodynamics and water vapor behavior. Warm air can hold more water vapor, increasing humidity. Relative humidity (RH) compares the current water vapor content to the maximum possible at a given temperature. Higher temperatures can lower RH if moisture remains constant, and vice versa. Dew point temperature is when air becomes saturated, forming dew. This relationship affects human comfort and weather, influencing phenomena like fog and precipitation. Understanding it aids in weather prediction and assessing environmental conditions. </H4>

```python
data.columns
```

### 4.3. Scatterplots and Correlation

```python
plt.figure(figsize=(15, 6))
plt.scatter(data['Temp_C'], data['Rel Hum_%'], color='skyblue')
plt.xlabel('Temperature')
plt.ylabel('Temperature')
plt.title('Scatter Plot of Two Continuous Variables')
plt.grid(True)
plt.show()
```

```python
corr_matrix = data[['Temp_C', 'Rel Hum_%']].corr()
plt.figure(figsize=(15, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 12})
plt.title('Correlation Heatmap between Temperature and Humidity')
plt.show()
```

## <hr class="solid"> 5. Relationship between a numerical and a categorical value <a id='sec5'></a>`

### 5.1. Values to be analyzed

<H4> 1. Temperature </br>
2. Weather </H4>

### 5.2. Science behind the relation between the numerival and categorical value

<h4> - Temperature influences the formation and behavior of weather phenomena such as clouds, precipitation, fog, thunderstorms, and winds. </br>
- Warm temperatures can lead to the evaporation of water from the Earth's surface, which contributes to the formation of clouds and precipitation when the moist air rises and cools. </br>
- Temperature gradients in the atmosphere can drive the movement of air masses, leading to changes in weather patterns and the development of wind systems. </h4>

### 5.3. Plotting the relationship

#### 5.3.1. Box Plot

```python
plt.figure(figsize=(20, 10))
sns.boxplot(data=data, x='Weather', y='Temp_C')
plt.xticks(rotation=45)
plt.title('Box Plot of Temperature by Weather')
plt.xlabel('Weather')
plt.ylabel('Temperature (Celsius)')
plt.xticks(rotation=90)
plt.show()
```

#### 5.3.2. Kernel Density Plot

```python
plt.figure(figsize=(25, 15))
sns.kdeplot(
    data=data,
    x="Temp_C",
    hue="Weather",
    common_norm=False
)
plt.title("Kernel Density Plot - Numerical Value by Category")
plt.xticks(rotation=90)
plt.show()
```
