## 1. Housekeeping

```python
# Importing necessary libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import constants
```

```python
# Reading input file

data = pd.read_csv('Weather Data.csv')
data
```

## 2. Analyzing Continuous Values

### 2.1. Analyzing the Temperature

```python
# Understanding the frequency distribution of the Temperature recored

plt.figure(figsize=(8, 6))
plt.hist(data['Temp_C'], bins=20, color='green', alpha=0.7)
plt.xlabel('Tempareture')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.tight_layout()
plt.show()
```

### 2.2. Analyzing the Wind Speed

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Wind Speed_km/h'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Wind Speed')
plt.ylabel('Frequency')
plt.title('Wind Speed Distribution')
plt.tight_layout()
plt.show()
```
